```python
# Impor library yang diperlukan
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import export_graphviz
import graphviz
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Memuat dataset Iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Memisahkan dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Membuat model Decision Tree
model = DecisionTreeClassifier()
```

```python
# Melatih model dengan data latih
model.fit(X_train, y_train)
```

```python
# Memprediksi kelas untuk data uji
y_pred = model.predict(X_test)
```

```python
# Mengukur akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Menampilkan laporan klasifikasi
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Prediksi kelas untuk instance baru
new_instance = [[5.1, 3.5, 1.4, 0.2]]  # Contoh instance baru
predicted_class = model.predict(new_instance)
print("Predicted Class:", predicted_class)
```

```python
# Export Decision Tree ke format dot
dot_data = export_graphviz(model, out_file=None,
                     feature_names=iris.feature_names,
                     class_names=iris.target_names,
                     filled=True, rounded=True,
                     special_characters=True)
```

```python
# Tampilkan visualisasi Decision Tree
graph = graphviz.Source(dot_data)
graph.render("iris_decision_tree", format='png', cleanup=True)
graph
```

```python
# Membuat confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Menampilkan confusion matrix menggunakan heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g',
            xticklabels=iris.target_names,
            yticklabels=iris.target_names)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
